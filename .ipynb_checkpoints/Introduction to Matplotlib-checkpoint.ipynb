## Matplotlib example workflow

```python
# 0. import matplotlib and get ready for plotting Jupyter 
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # (width, height) -> (10, 10) means the dimensions of this plot i.e. height and width

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show(you save the whole figure)
fig.savefig("new-images/sample-plot.png")
```

## Making figures with NumPy arrays

We want: 
* Line Plot
* Scatter Plot
* Bar Plot
* Hist
* Subplots

```python
import numpy as np
```

```python
# create some data
x = np.linspace(0, 10, 100) #Return evenly spaced numbers over a specified interval
x[:10]
```

#### Line Plot

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

#### Scartter Plot

```python
# use the same data to make a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x)); # the result is a plot of dots
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x)); # the result is a sin wave
```

#### bar 
* vertical bar
* horizontal bar

```python
# Make a plot from a dictionary
nut_butter_prices = {"Almond butter": 10, 
                     "Peanut butter": 8, 
                     "Cashew butter": 12}
fix, ax = plt.subplots()
# a bar plot has x and height not x and y axis
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(
    title="Leo's Nut Butter Store",
    ylabel="Price ($)"
);
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values())); # convert the dictioneries into a list to avoid an error
```

#### Histogram

```python
# make some data 
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

#### Subplots (Multiple-plots)
has two options 


```python
# Subplot option 1

# we have 4 figs/plots on 2 rows and 2 cols.
# the figs/plots have a width of 10 and height of 5
fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                           figsize=(10, 5))

# Plot to each different aixs
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2

fig, ax = plt.subplots(nrows=2,
                       ncols=2, 
                       figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2); # plots on row 0 and col 0
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

# Plotting from pandas DataFrames

```python
import pandas as pd
```

```python
# Make a dataframe
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

#### example plot from a series

```python
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range("1/1/2020", periods=1000))
ts = ts.cumsum()
ts.plot()
```

#### example from a dataframe

```python
car_sales
```

```python
# car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
# the above does not work on this notebook

car_sales["Price"] = (car_sales["Price"].replace('[\$\,\.]', '', regex=True).astype(float)/100).astype(int)
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
# remember Price is a string 
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# Plot the total Sales over time
car_sales.plot(x="Sale Date", y="Total Sales");
```

#### scatter plot

```python
# price is already int from above
# scatter plot is on odometer vs price
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```
