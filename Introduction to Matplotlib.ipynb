## Matplotlib example workflow

```python
# 0. import matplotlib and get ready for plotting Jupyter 
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # (width, height) -> (10, 10) means the dimensions of this plot i.e. height and width

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show(you save the whole figure)
fig.savefig("new-images/sample-plot.png")
```

#### Subplots (Multiple-plots)
has two options 


```python
# Subplot option 1

# we have 4 figs/plots on 2 rows and 2 cols.
# the figs/plots have a width of 10 and height of 5
fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                           figsize=(10, 5))

# Plot to each different aixs
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2

fig, ax = plt.subplots(nrows=2,
                       ncols=2, 
                       figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2); # plots on row 0 and col 0
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

#### Which one should you use? (pyplot vs matplotlib OO method?)
* There is no definitive answer
* When plotting something quickly use the pyplot method
* When plotting something more advanced, use the OO method

## the OO method

```python
heart_disease.head()
```

```python
# create a new dataset of people over 50
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# pyplot method
over_50.plot.scatter(x="age", y="chol", c="target"); # c=color
```

#### OOP method mixed with pyplot method

```python
# OO method of the above
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);

# ax.set_xlim([45, 100])
# changes/adjusts limit from (50 to 75) to (45 to 100)
```

#### OOP method from scratch

```python
# we're going to recreate the above from OOP straight an customize
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target"); # the asteriks(*) is unpacking the scatter

# Add a horizontal line across the axis
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

```python
over_50.head()
```

```python
# subplots of chol, age, thalch

fig, (ax0, ax1) = plt.subplots(nrows=2, 
                               ncols=1,
                               figsize=(10, 10), 
                               sharex=True);

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        # xlabel="Age", # after setting "sharex=True" you can remove this line
        ylabel="Cholesterol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Max Heat Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax1.axhl ine(y=over_50["thalach"].mean(),
            linestyle="--")

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# Set the style
plt.style.use("seaborn-v0_8-whitegrid")

## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter") # this changes the color scheme

# customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target"); # the asteriks(*) is unpacking the scatter

# Add a horizontal line across the axis
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

```python
# customizing the y and x axis limitations

# subplots of chol, age, thalch
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                               ncols=1,
                               figsize=(10, 10), 
                               sharex=True);

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter")

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        # xlabel="Age", # after setting "sharex=True" you can remove this line
        ylabel="Cholesterol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# ***************** change x axis limits here *****************
ax0.set_xlim([50, 80])

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="winter")

# Customize ax1
ax1.set(title="Heart Disease and Max Heat Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# ***************** change axis limits here *****************
ax1.set_xlim([50, 80]) # this is not necessary, both share the x-axis
ax1.set_ylim([60, 200])

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--")

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
